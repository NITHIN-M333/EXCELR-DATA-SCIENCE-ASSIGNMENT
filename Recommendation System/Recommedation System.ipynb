## Creating pivot table

```python
book_pivot=pd.pivot_table(data=bookcp,values='Book.Rating',index='User.ID',columns='Book.Title')
book_pivot
```

## Return the number of missing values in each columns

```python
book_pivot.isna().sum()
```

## Filling the NaN values with zeros

```python
book_pivot.fillna(0,inplace=True)
book_pivot.head()
```

## Model Building

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
rating_rate=1-pairwise_distances(X=book_pivot,metric='cosine')
rating_rate
```

```python
rating_df=pd.DataFrame(rating_rate)
rating_df
```

```python
rating_df.index=bookcp['User.ID'].unique()
rating_df.columns=bookcp['User.ID'].unique()
```

```python
rating_df
```

```python
user_50=rating_df.iloc[:50,:50]
user_50
```

## Filling the diagonal values of first 50 users as zeros

```python
np.fill_diagonal(user_50.to_numpy(),val=0)
```
