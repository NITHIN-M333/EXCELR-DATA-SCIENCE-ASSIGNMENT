```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
movies=pd.read_csv('my_movies.csv')
movies
```

```python
moviescp=movies.copy()
moviescp
```

```python
moviescp.head()
```

## Droping unwanted columns

```python
moviescp.drop(columns=['V1','V2','V3','V4','V5'],axis=1,inplace=True)
```

```python
moviescp
```

## Apriori

```python
frequent_item=apriori(moviescp,min_support=0.5,use_colnames=True)
frequent_item
```

## Association rules

```python
rule=association_rules(frequent_item,metric='confidence',min_threshold=0.8)
rule
```

```python
# tweeking the values of support and confidence
frequent_item2=apriori(moviescp,min_support=0.3,use_colnames=True)
frequent_item2
```
