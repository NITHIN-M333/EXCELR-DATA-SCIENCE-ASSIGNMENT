# ASSOCIATION RULES

## Apriori

```python
frequent_items=apriori(books,min_support=0.2,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='confidence',min_threshold=0.5)
rules
```

## Apriori 1

```python
# Tweeking the values of support and confidence
frequent_item_1=apriori(books,min_support=0.1,use_colnames=True)
frequent_item_1
```

```python
rules_1=association_rules(frequent_item_1,metric='confidence',min_threshold=0.4)
rules_1
```

## Apriori _2

```python
frequent_item_2=apriori(books,min_support=0.09,use_colnames=True)
frequent_item_2
```
